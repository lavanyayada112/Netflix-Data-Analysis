#Analyzing the top 5 directors

```python
df['director'] = df['director'].fillna('Director not  Specified')
df.head()
```

```python
directors_list = pd.DataFrame()
print(directors_list)
```

```python
directors_list = df['director'].str.split(',', expand=True).stack()
print(directors_list)
```

```python
directors_list = directors_list.to_frame()
print(directors_list)
```

```python
directors_list.columns = ['director']
print(directors_list)
```

```python
directors = directors_list.groupby(['director']).size().reset_index(name=' Total counts')
print(directors)
```

```python
directors = directors[directors.director != 'Director not Specified']
```

```python
print(directors)
```

```python
directors = directors.sort_values(by=[' Total counts'],ascending = False)
print(directors)
```

```python
top5Directors = directors.head()
print(top5Directors)
```

```python
top5Directors = top5Directors.sort_values(by=[' Total counts'])
barChart = px.bar(top5Directors, x=' Total counts', y='director', title='Top 5 Directors on Netflix')
barChart.show()
```

#Analyzing the top 5 acting on Netflix

```python
df['cast'] = df['cast'].fillna('No cast Specified')
cast_df = pd.DataFrame()
cast_df = df['cast'].str.split(',', expand=True).stack()
cast_df = cast_df.to_frame()
cast_df.columns = ['Actor']
actors = cast_df.groupby(['Actor']).size().reset_index(name='Total Counts')
actors = actors[actors.Actor != 'No cast Specified']
actors = actors.sort_values(by=['Total Counts'], ascending=False)
top5Actors = actors.head()
top5Actors = top5Actors.sort_values(by=['Total Counts'])
barChart = px.bar(top5Actors, x='Total Counts', y='Actor', title='Top 5 Actors on Netflix')
barChart.show()
```

#Analyzing the content produced on netflix based on years



```python
df1 = df[['type','release_year']]
df1=df1.rename(columns = {"release_year":"Release Year", "type": "type"})
df2=df1.groupby(['Release Year','type']).size().reset_index(name='Total Content')

```

```python
print(df2)
```

```python
df2 = df2[df2['Release Year']>=2000]
graph = px.line(df2, x="Release Year", y="Total Content", color="type", title="Trend of content produced on Netflix by Every Year")
graph.show()
```

#Sentiment Analysis of  netflix Content

```python
df3 = df[['release_year','description']]
df3 = df3.rename(columns={"release_year":"Release Year", "description":"Description"})

for index, row in df3.iterrows():
    d = row['Description']
    testimonial = TextBlob(d)
    p = testimonial.sentiment.polarity

    if p == 0:
        sent = 'Neutral'
    elif p > 0:
        sent = 'Positive'
    else:
        sent = 'Negative'

    df3.loc[index, 'Sentiment'] = sent

df3 = df3.groupby(['Release Year', 'Sentiment']).size().reset_index(name='Total Content')

df3 = df3[df3['Release Year'] >= 2005]

barGraph = px.bar(df3, x='Release Year', y='Total Content', color='Sentiment',
                  title='Sentiment Analysis of Netflix Content')
barGraph.show()
```
